```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("einstein.jpg", 0)
```

```python
imgEqualizada = cv2.equalizeHist(img)
```

```python
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(121) #1 linha 2 colunas referente a primeira img
plt.imshow(img, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgEqualizada, cmap=plt.cm.gray)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(121) #1 linha 2 colunas referente a primeira img
plt.hist(img.ravel(), 256, [0, 256])

ax2 = fig.add_subplot(122)
plt.hist(imgEqualizada.ravel(), 256, [0, 256])
plt.show()
```
