```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
imgOriginal = cv.imread("einstein.jpg", 0)
```

```python
totalLinhas, totalColunas = imgOriginal.shape
print(totalLinhas)
print(totalColunas)
```

```python
mat = cv.getRotationMatrix2D((totalLinhas/2, totalColunas/2), 45, 1) # 45 graus, side escala=1
```

```python
imgRotacionada = cv.warpAffine(imgOriginal, mat, (totalColunas, totalLinhas))
```

```python
fig = plt.figure(figsize=(10, 8))
plt.imshow(imgRotacionada, cmap = plt.cm.gray)
```
